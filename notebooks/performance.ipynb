```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Number of competitions participated vs average 3x3x3 cube means

compsAverages333 = pd.read_csv("../datasets/compsAverages333.csv")

groupedMeans = compsAverages333.groupby(by=['comps']).mean()
head = groupedMeans[:29]
tail = pd.DataFrame({"compAVG": groupedMeans[30:].mean()[0]}, index=[30])
combinedGroupedMeans = pd.concat([head, tail])

altIndex = [str(x) for x in range(1, 30, 2)]
altIndex.append("30+")
# combinedGroupedMeans.index = altIndex

ax = combinedGroupedMeans.plot()
ax.set_xticks([1, 5, 10, 15, 20, 25, 30])
ax.set_xticklabels([str(x) for x in [1, 5, 10, 15, 20, 25]] + ["30+"])

```
