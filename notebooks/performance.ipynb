```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Number of competitions participated vs average 3x3x3 cube means

compsAverages333 = pd.read_csv("../datasets/compsAverages333.csv")

groupedMeans = compsAverages333.groupby(by=['comps']).mean()
head = groupedMeans[:29]
tail = pd.DataFrame({"compAVG": groupedMeans[30:].mean()[0]}, index=[30])
combinedGroupedMeans = pd.concat([head, tail])

improvement = ((combinedGroupedMeans.iloc[0] - combinedGroupedMeans.iloc[29]) / combinedGroupedMeans.iloc[0])[0]
print("Time improvement between people that have competed in a single competition and those that have competed in 30 or more:")
print(f"{np.round(((combinedGroupedMeans.iloc[0] - combinedGroupedMeans.iloc[29]) / combinedGroupedMeans.iloc[0])[0] * 100, 2)}%")

altIndex = [str(x) for x in range(1, 30, 2)]
altIndex.append("30+")

ax = combinedGroupedMeans.plot()
ax.set_xticks([1, 5, 10, 15, 20, 25, 30])
ax.set_xticklabels([str(x) for x in [1, 5, 10, 15, 20, 25]] + ["30+"])

```

```python
# Correlation between 3x3x3 and 4x4x4 performance

corr333_444 = pd.read_csv("../datasets/333vs444.csv")
print(len(corr333_444))

avg333mean = corr333_444.mean()['333']
avg444mean = corr333_444.mean()['444']
avg333std = corr333_444.std()['333']
avg444std = corr333_444.std()['444']

for i in range(len(corr333_444)):
    if abs(corr333_444['333'][i] - avg333mean) > 1 * avg333std:
        corr333_444 = corr333_444.drop(i)
        continue
    if abs(corr333_444['444'][i] - avg444mean) > 1 * avg444std:
        corr333_444 = corr333_444.drop(i)

corr333_444['333'] = (max(corr333_444['333']) - corr333_444['333']) / (max(corr333_444['333']) - min(corr333_444['333']))
corr333_444['444'] = (max(corr333_444['444']) - corr333_444['444']) / (max(corr333_444['444']) - min(corr333_444['444']))
```

```python
corr333_444_Sample =  corr333_444.sample(250, random_state=1)
plt.rcParams['figure.figsize'] = [8, 8]
plt.scatter(corr333_444_Sample['333'], corr333_444_Sample['444'])
plt.xlabel('333')
plt.ylabel('444')
```
