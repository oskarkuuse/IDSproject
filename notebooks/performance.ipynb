```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
compParticipations = pd.read_csv("../datasets/compsParticipated.csv")
compParticipationsFreq = compParticipations.value_counts().sort_index()
compParticipationsFreq.index = list(x[0] for x in compParticipationsFreq.index)

compParticipationsFreq[:10].plot(kind='bar')
```

```python
compsAverages333 = pd.read_csv("../datasets/compsAverages333.csv")
groupedMeans = compsAverages333.groupby(by=['comps']).mean()
groupedMeans.plot()

```
